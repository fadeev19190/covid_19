```python
import pandas as pd
import plotly.express as px
from dash import dcc, html
from dash.dependencies import Output, Input
from jupyter_dash import JupyterDash

covid = (
        pd.read_csv('WHO-COVID-19-global-data.csv', parse_dates=['Date_reported'])
        .set_index('Date_reported'))

covid.head(10)
```

```python
covid_date = covid.groupby('Date_reported')['New_cases'].sum().reset_index()

covid_date['Date_reported'] = pd.to_datetime(covid_date['Date_reported'])

covid_date.set_index('Date_reported', inplace=True)

options_dict = {
    'D': 'Daily',
    'M': 'Monthly',
    'Q': 'Quarterly',
    'A': 'Annual'
}
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='dropdown',
        options=options_dict,
        value='D',
    ),
    dcc.Graph(id='graph'),
])
@app.callback(Output('graph', 'figure'), Input('dropdown', 'value'))
def line_chart(resample):
    fig_cases = px.line(
        covid_date.resample(resample).mean(),
        title=f"{options_dict[resample]} Covid New Cases",
    ).update_layout(yaxis_title='New_cases', showlegend=False)
    
    
    return fig_cases

if __name__ == "__main__":
    app.run_server(mode='external')
```
